```python
# Load env variables

from dotenv import load_dotenv
load_dotenv()
```

```python
# Create an API client

from anthropic import Anthropic

client = Anthropic()
model = "claude-sonnet-4-6"
```

```python
# Creating helper functions

# Helper functions to add user and assistant messages

def add_user_message(messages, text):
    user_message = {"role": "user", "content": text}
    messages.append(user_message)

def add_assistant_message(messages, text):
    assistant_message = {"role": "assistant", "content": text}
    messages.append(assistant_message)
```

```python
# Helper function chat (001_request) updated to stream Claude's responses

messages = []

add_user_message(messages, "Write a 1 sentence description of a fake database")

# Streaming the messages received from Claude
stream = client.messages.create(
    model=model,
    max_tokens=1000,
    messages=messages,
    stream=True
)
for event in stream:
    print(event)
```

```python
# Helper function using the stream function of Anthropic SDK to stream Claude's responses

messages = []

add_user_message(messages, "Write a 1 sentence description of a fake database")

with client.messages.stream(
    model=model,
    max_tokens=1000,
    messages=messages
) as stream:
    for text in stream.text_stream:
        print(text, end="")
```
