```python
# Create an API client

from anthropic import Anthropic

client = Anthropic()
model = "claude-sonnet-4-6"
```

```python
# Make a request

# model - The name of the Claude model you want to use
# max_tokens - A safety limit on response length (not a target)
# messages - The conversation history you're sending to Claude

message = client.messages.create(
    model=model,
    max_tokens=1000,
    messages=[
        {
            "role": "user",
            "content": "What is generative AI? Answer in one sentence"
        }
    ]
)

# Claude doesn't store any of your conversation history. Each request you make is completely independent, with no memory of previous exchanges.

message = client.messages.create(
    model=model,
    max_tokens=1000,
    messages=[
        {
            "role": "user",
            "content": "Write another sentence"
        }
    ]
)
```

```python
# User messages - Content you want to send to Claude (written by humans)
# Assistant messages - Responses that Claude has generated

message
```

```python
# Creating helper functions

def add_user_message(messages, text):
    user_message = {"role": "user", "content": text}
    messages.append(user_message)

def add_assistant_message(messages, text):
    assistant_message = {"role": "assistant", "content": text}
    messages.append(assistant_message)

def chat(messages):
    message = client.messages.create(
        model=model,
        max_tokens=1000,
        messages=messages,
    )
    return message.content[0].text
```

```python
# Maintaining a conversation using the helper functions

# Start with an empty message list
messages = []

# Add the initial user question
add_user_message(messages, "What is Generative AI. Answer in one sentence")

# Get Claude's response
answer = chat(messages)

# Add Claude's response to the conversation history
add_assistant_message(messages, answer)

# Add a follow-up question
add_user_message(messages, "Write another sentence")

# Get the follow-up response with full context
final_answer = chat(messages)
```
