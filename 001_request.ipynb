```python
# Create an API client

from anthropic import Anthropic

client = Anthropic()
model = "claude-sonnet-4-6"
```

```python
# Make a request

# model - The name of the Claude model you want to use
# max_tokens - A safety limit on response length (not a target)
# messages - The conversation history you're sending to Claude

message = client.messages.create(
    model=model,
    max_tokens=1000,
    messages=[
        {
            "role": "user",
            "content": "What is generative AI? Answer in one sentence"
        }
    ]
)

# Claude doesn't store any of your conversation history. Each request you make is completely independent, with no memory of previous exchanges.

message = client.messages.create(
    model=model,
    max_tokens=1000,
    messages=[
        {
            "role": "user",
            "content": "Write another sentence"
        }
    ]
)
```

```python
# Creating helper functions

def add_user_message(messages, text):
    user_message = {"role": "user", "content": text}
    messages.append(user_message)

def add_assistant_message(messages, text):
    assistant_message = {"role": "assistant", "content": text}
    messages.append(assistant_message)

# # Function without system prompt providing a guidance on how Claude must respond
# def chat(messages):
#   message = client.messages.create(
#        model=model,
#        max_tokens=1000,
#       messages=messages,
#   )
#   return message.content[0].text

# function with system prompt and later added also the temperature

def chat(messages, system=None, temperature=1.0):
    params = {
        "model": model,
        "max_tokens": 1000,
        "messages": messages,
        "temperature": temperature
    }
    
    if system:
        params["system"] = system
    
    message = client.messages.create(**params)
    return message.content[0].text
```

```python
# Maintaining a conversation using the helper functions

# Start with an empty message list
messages = []

# Add the initial user question
add_user_message(messages, "What is Generative AI. Answer in one sentence")

# Get Claude's response
answer = chat(messages)

# Add Claude's response to the conversation history
add_assistant_message(messages, answer)

# Add a follow-up question
add_user_message(messages, "Write another sentence")

# Get the follow-up response with full context
final_answer = chat(messages)
```

```python
# Answer without system prompt providing a guidance on how Claude must respond
messages = []

add_user_message(messages, "How do I solve 5x + 2 = 3 for x?")

# Get Claude's response
answer = chat(messages)

print(answer)
```

```python
# Answer with system prompt providing a guidance on how Claude must respond
messages = []

# With system prompt
system = """
You are a patient math tutor.
Do not directly answer a student's questions.
Guide them to a solution step by step.
"""

add_user_message(messages, "How do I solve 5x + 2 = 3 for x?")

# Get Claude's response
answer = chat(messages, system=system)

print(answer)
```

```python
# Answer with different temperatures
messages = []

add_user_message(messages, "Generate a one sentence movie idea")

# Get Claude's response
answer = chat(messages, temperature=1.0)

print(answer)
```
